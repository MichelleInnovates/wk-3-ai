```python
# Deliverable for Part 2, Task 1

import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
from sklearn.preprocessing import LabelEncoder

# 1. Load the Dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

print("Dataset loaded successfully.")

# 2. Preprocess the data
# The Iris dataset is clean and has no missing values, so we can proceed.
# Labels are already numeric (0, 1, 2), so encoding is not strictly necessary,
# but we show the step for completeness if they were strings.
le = LabelEncoder()
y_encoded = le.fit_transform(y) # This won't change y in this case but is good practice

# 3. Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)
print(f"Data split into {len(X_train)} training samples and {len(X_test)} testing samples.")

# 4. Train a Decision Tree Classifier
# Initialize the model
clf = DecisionTreeClassifier(random_state=42)

# Train the model on the training data
clf.fit(X_train, y_train)
print("Decision Tree Classifier trained.")

# 5. Make predictions on the test set
y_pred = clf.predict(X_test)

# 6. Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
# Use 'macro' average for precision and recall to treat all classes equally
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')

print("\n--- Model Evaluation ---")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision (Macro): {precision:.4f}")
print(f"Recall (Macro): {recall:.4f}")

# Display a detailed classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=target_names))
```
